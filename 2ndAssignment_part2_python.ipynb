```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)

n=int(input('Enter a number : '))
if n<0:
    print('Factorial of a negative number is not possible')
elif n==0:
    print('Factorial of 0 = 1')
else:
    a=factorial(n)
    print(f"Factorial of {n} = {a}")
```

```python
board = [' ' for _ in range(9)]

def print_board():
    print(" " + board[0] + " | " + board[1] + " | " + board[2])
    print("---+---+---")
    print(" " + board[3] + " | " + board[4] + " | " + board[5])
    print("---+---+---")
    print(" " + board[6] + " | " + board[7] + " | " + board[8])

def check_win(player):
    win_combinations = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
                        (0, 3, 6), (1, 4, 7), (2, 5, 8),
                        (0, 4, 8), (2, 4, 6)]

    for combo in win_combinations:
        if all(board[i] == player for i in combo):
            return True
    return False

def check_draw():
    return ' ' not in board

current_player = 'X'

while True:
    print_board()
    print(f"Player {current_player}'s turn. Enter a position (1-9):")

    while True:
        try:
            move = int(input())
            if 1 <= move <= 9 and board[move - 1] == ' ':
                break
            else:
                print("Invalid input. Try again.")
        except ValueError:
            print("Invalid input. Try again.")

    board[move - 1] = current_player

    if check_win(current_player):
        print_board()
        print(f"Player {current_player} wins!")
        break
    elif check_draw():
        print_board()
        print("It's a draw!")
        break

    current_player = 'O' if current_player == 'X' else 'X'

    
```

```python
def prime_factors(n):
    fac = []
    div = 2
    while div <= n:
        if n % div == 0:
            fac.append(div)
            n //= div
        else:
            div += 1
    return fac
n=int(input('Enter a number : '))
fac=prime_factors(n)
print ('Prime factors of ',n,' are : ',end="")
for i in fac:
    print(i,' ',end="")
```
